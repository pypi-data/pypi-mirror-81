## Plot example gene: DNMT3B

```python
import brie
import numpy as np

genes = brie.loadgene("/home/yuanhua/BitRepo/thesis/data/SLC25A3_use.GFF")
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

from pyseqlib.plot import plot_gene, set_frame

import hilearn
hilearn.plot.set_style()

import seaborn
color4 = seaborn.color_palette("hls", 4)
color8 = seaborn.color_palette("hls", 8)


fig = plt.figure(figsize=(14,4))
ax = plt.subplot(1,2,1)
plot_gene(genes[0], tran_colors=color8)
plt.title("SLC25A3")
set_frame(ax)

ax = plt.subplot(1,2,2)
exon_emph = [[98595433, 98595557],[98595727,98595848]]
exon_colors = [color4[0], color4[1]]
plot_gene(genes[0], exon_emph=exon_emph, exon_colors=exon_colors)
plt.title("SLC25A3")
set_frame(ax)

# fig.savefig("/home/yuanhua/Pictures/SLC25A3.pdf",dpi=300, bbox_inches='tight')
plt.show()
```
